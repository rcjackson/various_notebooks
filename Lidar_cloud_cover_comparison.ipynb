```python
import sage_data_client
import matplotlib.pyplot as plt

```

```python
df = sage_data_client.query(start="-12h",
    filter={"name": "weather.classifier.class"}
)
df = df.set_index('timestamp')
df['value'].plot(label='lidar flag (0 = clear, 1 = cloud)')
# query and load data into pandas data frame
df_camera = sage_data_client.query(
    start="-12h",
    filter={
        "name": "env.coverage.cloud",
    }
)

df_camera = df_camera.set_index('timestamp')
df_camera['value'].plot(label='camera coverage')
plt.legend()
plt.ylabel('Cloud cover fraction')
```
